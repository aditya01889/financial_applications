```python
from time import sleep
import sys

def bond_val():
    
    global face_value
    global coupon_rate
    global discount_rate
    global maturity
    global bond_freq
    global discount_freq
    
    print("\nEnter only NUMBERS and in DECIMAL FORM when necessary. E.g. 20% --> 0.2")
    while True:
        try:
            face_value = int(((input("What is the face value? "))))
            break
        except:
            print("Please enter a whole number!")
    while True:
        try:
            coupon_rate = float(input("What is the coupon rate p/a? "))
            break
#         else:
#             coupon_rate = input("What is the coupon rate p/a?")
#             if coupon_rate[-1] == "%":
#                 coupon_rate = float(coupon_rate[:-1])/100
        except:
            print("Please enter a number!")
    while True:
        try:
            discount_rate = float((input("What is the discount rate p/a? ")))
            break
#         else:
#             discount_rate = input("What is the discount rate p/a?")
#             if discount_rate[-1] == "%":
#                 discount_rate = float(discount_rate[:-1])/100
#                 break
        except:
            print("Please enter a number!")
    while True:
        try:
            maturity = int((input("How many years until this bond reaches maturity? ")))
            break
        except:
            print("Please enter a whole number!")
    
    while True:
        try:
            more_bond_options = str(input("Are both rates semi-annual? Y/N "))
            break
        except:
            print("Please enter Y or N!")
    
    if more_bond_options.lower() == "n":
        while True:
            try:
                bond_freq = int(input("How many times does the bond pay out per year? "))
                break
            except:
                print("Please enter a whole number!")
        while True:
            try:
                discount_freq = int(input("How many times does the discount rate compound per year? "))
                break
            except:
                print("Please enter a whole number!")        
    else:
        print("You have selected Yes.")
        bond_freq = 2
        discount_freq = 2
                
def bond_calc():
    
    global face_value
    global coupon_rate
    global discount_rate
    global maturity
    global bond_freq
    global discount_freq
    
    coupon_rate = coupon_rate/bond_freq
    maturity = maturity*bond_freq
    discount_rate = ((1+discount_rate/discount_freq)**(discount_freq/bond_freq))-1
    
    coupon = coupon_rate*face_value
    annuity_factor = (1-(1+discount_rate)**(maturity*-1))/discount_rate
    pay = coupon*annuity_factor
    pv_fv = face_value/(1+discount_rate)**maturity
    pv = pay + pv_fv
    print("\nTHE BOND IS WORTH ${:.2f}".format(round(pv, 2)))
```

```python
def share_const_gr_model():
    print("Enter only NUMBERS and in DECIMAL FORM when necessary. E.g. 20% --> 0.2")
    while True:
        try:
            last_dividend = float(input("What was the last dividend paid? "))
            break
        except:
            print("Please enter a number.")
    while True:
        try:
            div_gr_rate = float(input("What is the growth rate of the dividend? "))
            break
        except:
            print("Please enter a number.")
    while True:
        try:
            discount_rate = float(input("What is the discount rate? "))
            break
        except:
            print("Please enter a number.")
    
    if discount_rate < div_gr_rate:
        print("Error: Discount Rate larger than Growth Rate.\nReporting to ASIC")
        sleep(1)
        print(".")
        sleep(2)
        print("..")
        sleep(3)
        print("...")
        sleep(1)
        sys.exit()
    
    share_value = (last_dividend*(1+div_gr_rate))/(discount_rate - div_gr_rate)
    print("SHARE PRICE is ${:.2f}".format(round(share_value, 2)))
```

```python
def share_two_dcf_model():
    print("Enter only NUMBERS and in DECIMAL FORM when necessary. E.g. 20% --> 0.2")
    first_stage_terms_list = []
    total_dividends = 0
    while True:
        try:
            first_stage_terms = int(input("How many periods in the first stage? "))
            break
        except:
            print("Please enter a whole number.")
    while True:
        try:
            discount_rate = float(input("What is the discount rate? "))
            break
        except:
            print("Please enter a number.")
    while True:
        try:
            div_gr_rate = float(input("What is the growth rate in the second stage? "))
            break
        except:
            print("Please enter a number.")
    
    for term in range(1, first_stage_terms+1):
        term_div = float(input("What is the value of the dividend in period {}? ".format(term)))
        first_stage_terms_list.append(term_div)
    
    for i in range(1, len(first_stage_terms_list)+1):
        discounted_dividend = (first_stage_terms_list[i-1])/((1 + discount_rate)**i)
        total_dividends += discounted_dividend
        discounted_dividend = 0
    
    future_share_value = ((first_stage_terms_list[-1]*(1+div_gr_rate))/(discount_rate - div_gr_rate)/
        (1+discount_rate)**(len(first_stage_terms_list)))
    
    share_value = total_dividends + future_share_value
    
    print("\nSHARE PRICE IS ${:.2f}".format(round(share_value, 2)))
    
                
```
